```python
import numpy as np
import pandas as pd
```

```python
df_data= pd.read_csv('Results.csv', index_col=0)
df_data
# read the data
```

```python
pd.DataFrame(df_data,columns=['n_participants','grouping','mean_watscar','n_participants','diet_group','age_group'])
#choose the needed data
```

```python
df_data["Total_consum"] = df_data['mean_acid'] * df_data['n_participants']

df_data
#calculate the total consumption
```

```python

df_sum=df_data.groupby(['diet_group','age_group'],as_index=False).sum()
# regroup the data by the same diet group and age group
df_sum['Aveage'] = df_sum['Total_consum'] /df_sum['n_participants']
df_sum
#calculate the average number of each group
```
